```python
from datetime import date
from IPython.display import clear_output
from time import sleep

class carpooling:
    # class variable for recording users information
    total = [['9722743101', 10, 55, 'p', date(2023,  12,  1)],
             ['9722743102', 23, 87, 'p', date(2023,  12,  1)],
             ['9722743103', 16, 97, 'p', date(2023,  12,  1)],
             ['9876796104', 77, 98, 'p', date(2023,  12,  1)],
             ['9722743105', 45, 88, 'p', date(2023,  12,  1)],
             ['9734226106', 89, 89, 'p', date(2023,  12,  1)],
             ['9936467107', 90, 90, 'p', date(2023,  12,  1)],
             ['9712743108', 22, 64, 'p', date(2023,  12,  1)],
             ['9943867109', 43, 41, 'p', date(2023,  12,  1)],
             ['9722743110', 65, 55, 'b', date(2023,  12,  1)],
             ['9722743111', 14, 55, 'b', date(2023,  12,  1)]]

    drivers = []
    passengers = []
    b = []
    
    def __init__(self):
        pass
    
    # ================= Add request =================================================================================
    def _request_checker(self, stu_id, date):
        i = 1 
        for LIST in carpooling.total:
            if (stu_id in LIST) & (date in LIST):
                return False
            else:
                if i == len(carpooling.total):
                    return True
                else:
                    i += 1
                    continue
    
    def _stuid_checker(self, ID):
        if len(ID) == 10:
            try:
                ID = int(ID)
                return True
            except:
                return False
        else:
            return False
        
    def _coord_checker(self, coord):
        if len(coord) == 2:
            try:
                x, y = int(coord[0]), int(coord[1])
                cond1 = (x >= 10) & (x <= 99)
                cond2 = (y >= 10) & (y <= 99)
                if cond1 & cond2:
                    return True
                else:
                    return False
            except:
                return False
        else:
            return False
    
    def _time_checker(self, time):
        if (len(time) == 3):
            try:
                time = date(int(time[0]), int(time[1]), int(time[2]))
                if time < date.today():
                    return False
                else:
                    return True
            except:
                return False
        else:
            return False
    
    def _input_converter(self, string):
        LIST = string.replace("[","").replace("]","").split(',')
        if len(LIST) == 5:
            LIST[4] = LIST[4].split('-')
            return LIST
        else:
            raise ValueError
        
        
    def _set_request_grp(self):
        while True:
            choosen_requests = []
            ind_reqs = input('''\nEnter your information! your input must exactly determine the format below:
Format = "[[Student_id,X,Y,Type,Date],[Student_id,X,Y,Type,Date],[Student_id,X,Y,Type,Date],...]"
* Student_id : 10digits
* X,Y : between 10 and 99
* Type : "R" or "B" or "P"
* Date : "year-month-day" or "" to get the system date
===>''')
            # Convert the string into devided lists in one bigger list 
            ind_reqs = ind_reqs.replace('[[','').replace(']]','').split('],[')
            
            # Check if the number of inputs is higher that one
            if len(ind_reqs) <= 1:
                print('\n* You should enter more than one request!')
                continue
            else:
                pass
            
            for i in range(len(ind_reqs)):
                ind_reqs[i] = ind_reqs[i].split(',')
                ind_reqs[i][4] = ind_reqs[i][4].split('-')
                
                # Student id check
                if self._stuid_checker(ind_reqs[i][0]):
                    pass
                else:
                    print(f'\n* In request #{i+1} the student id of {ind_reqs[i][0]} is invalid!\n!')
                
                # Coordination check
                if self._coord_checker(ind_reqs[i][1:3]):
                    ind_reqs[i][1], ind_reqs[i][2] = int(ind_reqs[i][1]), int(ind_reqs[i][2])
                    pass
                else:
                    print(f'\n* In request #{i+1} the coordination of (x = {ind_reqs[i][1]}, y = {ind_reqs[i][2]}) is invalid!\n')
                    continue
                    
                # Type check
                if ind_reqs[i][3].lower() in ['p', 'd', 'b']:
                    ind_reqs[i][3] = ind_reqs[i][3].lower()
                    pass
                else:
                    print(f'\n* In request #{i+1} the type {ind_reqs[i][3]} is invalid!\n')
                    continue
                    
                # Date check
                if ind_reqs[i][4] == ['']:
                    time = date.today()
                    ind_reqs[i][4] = date(time.year, time.month, time.day+1)
                    pass
                elif self._time_checker(ind_reqs[i][4]):
                    ind_reqs[i][4] = date(int(ind_reqs[i][4][0]), int(ind_reqs[i][4][1]), int(ind_reqs[i][4][2]))
                    pass
                else:
                    print(f"\n* In request #{i+1} the date of {'-'.join(ind_reqs[i][4])} is invalid!\n")
                    continue
                    
                 # Request check: to ensure that there aren't more that one request from a specific student in one day
                if self._request_checker(ind_reqs[i][0], ind_reqs[i][4]):
                    print(f'The request #{i+1} is successfully added')
                    carpooling.total.append(ind_reqs[i])
                else:
                    print(f'\n* The student {ind_reqs[i][0]} has already set request in {ind_reqs[i][4]}!\n')
                    continue
            
            while True:
                click = input('Return to main menu: (Y/N)')
                if click.lower() == 'y':
                    return True
                elif click.lower() == 'n':
                    break
                else:
                    continue
                    
    def _set_request_ind(self):
        while True:
            ind_req = input('''\nEnter your information! your input must exactly determine the format below:
Format = "[Student_id,X,Y,Type,Date]"
* Student_id : 10digits
* X,Y : between 10 and 99
* Type : "R" or "B" or "P"
* Date : "year-month-day" or "" to get the system date
===>''')
            try:
                ind_req = self._input_converter(ind_req)
            except:
                print('\n* Invalid input! Try again!')
                continue
                
            # Student id check
            if self._stuid_checker(ind_req[0]):
                pass
            else:
                print(f'\n* Student id of {ind_req[0]} is invalid! Try again!\n')
                continue
                
            # Coordination check
            if self._coord_checker(ind_req[1:3]):
                ind_req[1], ind_req[2] = int(ind_req[1]), int(ind_req[2])
                pass
            else:
                print(f'\n* The coordination of (x = {ind_req[1]}, y = {ind_req[2]}) is invalid! Try again!\n')
                continue
                
            # Type check
            if ind_req[3].lower() in ['p', 'd', 'b']:
                ind_req[3] = ind_req[3].lower()
                pass
            else:
                print(f'\n* The Type {ind_req[3]} is invalid! Try again!\n')
                continue
                
            # Date check
            if ind_req[4] == ['']:
                time = date.today()
                ind_req[4] = date(time.year, time.month, time.day+1)
                pass
            elif self._time_checker(ind_req[4]):
                ind_req[4] = date(int(ind_req[4][0]), int(ind_req[4][1]), int(ind_req[4][2]))
                pass
            else:
                print(f"\n* The date of {'-'.join(ind_req[4])} is invalid! Try again!\n")
                continue
                
            # Request check: to ensure that there aren't more that one request from a specific student in one day
            if self._request_checker(ind_req[0], ind_req[4]):
                break
            else:
                print(f'\n* The student {ind_req[0]} has already set request in {ind_req[4]}! Try again!\n')
                continue
        carpooling.total.append(ind_req)       
        return None
    
    # ================= Edit request ================================================================================
    def _find_sort_request(self, req_id):
        filtered_list = []
        
        for i in range(len(carpooling.total)):
            if carpooling.total[i][0] == req_id:
                filtered_list.append(carpooling.total[i])
            else:
                pass
        filtered_list.sort(key = lambda x: x[4])
        return filtered_list 
        
    def _find_index_request(self, edit_list):
        indices = []
        for j in range(len(edit_list)):
            for i in range(len(carpooling.total)):
                if carpooling.total[i] == edit_list[j]:
                    indices.append([i, j])
                else:
                    pass        
        return indices
    
    def _table1(self,indices, edit_list):
        i = 0
        print("{:>4}{:>15}{:>4}{:>4}{:>4}{:>20}".format('Index','Student_id','X','Y','T','Date'))
        while i <= 2:
            try:
                print("{:>4}{:>15}{:>4}{:>4}{:>4}{:>20}".format(indices[i][0], edit_list[i][0], edit_list[i][1], edit_list[i][2],edit_list[i][3],
                                                                    edit_list[i][4].strftime("%Y-%m-%d")))
                i += 1
                continue
            except:
                break      
        return None
    
    def _edit_main(self, indx): 
        while True:
            choice = input('''Which attribute do you want to edit
(1) X,Y
(2) T
(3) date
Choice: ''')
            if choice in [f'{i}' for i in range(1,5)]:
                
                if choice == '1':
                    xy = input('X;Y = ').split(';')
                    if self._coord_checker(xy):
                        x,y = int(xy[0]), int(xy[1])
                        carpooling.total[indx][1], carpooling.total[indx][2] = x, y
                        break
                    else:
                        print('Try again!')
                        continue
                    
                elif choice == '2':
                    t = input('type(r,d,b) = ').lower()
                    if t in ['r', 'd', 'b']:
                        carpooling.total[indx][3] = t
                        answer = input('Any other changes ? (y/n)')
                        if answer == 'y':
                            continue
                        else:
                            break
                    else:
                        continue
                        
                elif choice == '3':
                    time = input('''Enter the intended time (format = year-month-day)!
if you do not want to enter a time press `0` bottom''')
                    time = time.split('-')
            
                    if time == ['0']:
                        time = date.today()
                        carpooling.total[indx][4] = time
                    elif self._time_checker(time):
                        time = date(int(time[0]), int(time[1]), int(time[2]))
                        if self._request_checker(carpooling.total[indx][0], time):
                            carpooling.total[indx][4] = time
                        else:
                            print(f'''The student with the id of {carpooling.total[indx][0]} has already set 
                            a request at {time}! please set another time!\n''')
                            continue
                    else:
                        continue
            else:
                print('Try again!')
                continue
        return None 
    
    def edit_request(self):
        while True:
            req_id = input('Enter the student id that you want to edit its information:\n')
            if req_id in [carpooling.total[i][0] for i in range(len(carpooling.total))]:
                edit_list = self._find_sort_request(req_id)
                indices = self._find_index_request(edit_list)
                # display the requests and their indices to be revised
                print(indices)
                display = self._table1(indices, edit_list)
                indxs = [indices[i][0] for i in range(len(indices))]
                
                while True:
                    indx = int(input('index = '))
                    if indx in indxs:
                        change = self._edit_main(indx)
                        break
                    else:
                        print('Try again')
                        continue
             
                return True
                               
            else:
                choice = input('''The entered student_id is not available or is incorrect! 
(1) try again
(2) return to main menu
choice = ''')
                if choice == '1':
                    continue
                elif choice == '2':
                    return False
               
                                     
    # ================= Allocation   ================================================================================
    def _build_alloc_list1(self):
        allocation_list = []
        if len(carpooling.total) == 0:
            pass
        else:
            unique_dates = list(set([request[4] for request in carpooling.total]))
            for t in unique_dates:
                temp = []
                d,p,b = [],[],[]
                temp.append(t)
                for req in carpooling.total:
                    if req[4] == t:
                        if req[3] == 'd':
                            d.append(req[:3])
                        elif req[3] == 'p':
                            p.append(req[:3])
                        elif req[3] == 'b':
                            b.append(req[:3])
                    else:
                        continue
                temp.append(d)
                temp.append(p)
                temp.append(b)
                allocation_list.append(temp)      
        return allocation_list
    
    def _build_alloc_list2(self):
        allocation_list = self._build_alloc_list1()
        if len(allocation_list) == 0:
            pass
        else:
            for i in range(len(allocation_list)):
                while (4*len(allocation_list[i][1]) < (len(allocation_list[i][2]) + len(allocation_list[i][3]))):
                    if len(allocation_list[i][3]) != 0:
                        allocation_list[i][3].sort(key = lambda x: ((x[1] - 0)**2 + (x[2] - 0)**2)**0.5, reverse=True)
                        allocation_list[i][1].append(allocation_list[i][3].pop(0))
                    else:
                        break       
        return allocation_list
    
    def allocation(self):
        travels = []
        trip_id = 0
        allocation_list = self._build_alloc_list2()
        for i in range(len(allocation_list)):  # loop over dates
            while True:
                travel = []
                passenger = []
                capacity = 4
                if len(allocation_list[i][1]) != 0:
                    trip_id += 1
                    travel.append(trip_id)
                    travel.append(allocation_list[i][0].strftime("%Y-%m-%d"))
                    allocation_list[i][1].sort(key = lambda x: ((x[1] - 0)**2 + (x[2] - 0)**2)**0.5, reverse=True)
                    temp_driver = allocation_list[i][1].pop(0)
                    travel.append(temp_driver)
                    while capacity != 0:
                        if len(allocation_list[i][2]) != 0:
                            allocation_list[i][2].sort(key = lambda x: ((x[1] - temp_driver[1])**2 + (x[2] - temp_driver[2])**2)**0.5)
                            passenger.append(allocation_list[i][2].pop(0))
                            capacity -= 1
                            continue
                        elif len(allocation_list[i][3]) != 0:
                            allocation_list[i][3].sort(key = lambda x: ((x[1] - temp_driver[1])**2 + (x[2] - temp_driver[2])**2)**0.5)
                            passenger.append(allocation_list[i][3].pop(0))
                            capacity -= 1
                            continue
                        else:
                            break
                    # add list of passengers information to the current travel list    
                    travel.append(passenger)
                    travels.append(travel)
                else:
                    break
        return travels, allocation_list
                                     
    # ================= Add comment  ================================================================================
    def add_comment(self):           
        return None                  
    
    # ================= Edit comment ================================================================================
    def edit_comment(self):
        return None
```
