```python
def index_nonezero (array,lb,k):
    n = len(array)
    sum = 0
    for i in range (0,999999999999999):
        if (lb + i) < n:
            sum += (array[lb+i] != 0)
            if sum == k:
                break
            else:
                continue
            
        else:
            sum += (array[(lb + i)%n] != 0)
            if sum == k:
                break
            else:
                continue
    if (lb + i) < n:
        return (lb+i)
    else:
        return (lb+i)%n
```

```python
def Zero(a):
    if a <= 0:
        a = 0
        return a
    else:
        return a             
```

```python
def m_iter(asset, m, M,Iter):
    if asset[m-1] == 0:
        M.append(Iter)
        return min(M)
    else:
        M.append(10**3)
        return min(M)
```

```python
def breakdown(k,n,l,m):

#___m condition_____________________________________________
    
    if (m > n) and (m <= 0):
        print('invalid value of m!')
    else:
        pass
    
#___l condition_____________________________________________    
    
    if (l > 1) and (l <= 0):
        print('invalid value of m')
    else:
        pass
    
#___k condition______________________________________________    
    
    if k > n:
        return 'invalid value of k!'
    else:
        pass
#____________________________________________________________
    
    asset = np.ones((n))
    i = k - 1
    i = int(i)
    m = int(m)
    M = []
    asset[i] -= l
    Iter = 1
    get_status(asset,n,Iter)           

    while sum(asset) > 0:
        iterm = m_iter(asset, m, M,Iter)
        i+=k
        if i >= n:
            t = i%n
            remain_non_zero = k - ((sum(asset[i-k+1:] != 0))+(sum(asset[:(i%n)+1] != 0)))
            i = index_nonezero(asset,i+1, remain_non_zero)
            asset[i] -= l
            asset[i] = Zero(asset[i])
            Iter += 1
                
        else:
            remain_non_zero = k - ((sum(asset[i-k+1:i+1] != 0)))
            i = index_nonezero(asset,i+1, remain_non_zero)
            asset[i] -= l
            asset[i] = Zero(asset[i])
            Iter += 1
            
        get_status(asset,n,Iter)           
    print('Asset',m,'has been lost its capacity in iteration number',iterm)
    return None
```
